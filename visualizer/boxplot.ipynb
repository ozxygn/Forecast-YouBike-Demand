```python
# MAE

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the two files and tag their source (kept in df but not used for hue)
f1 = pd.read_csv(
    r'C:\Users\asus\Dokumen_Billi\allstationpredgit\result\metrics_all_stations5days.csv'
)
f1['Source'] = '5days All Stations'

f2 = pd.read_csv(
    r'C:\Users\asus\Dokumen_Billi\allstationpredgit\result\lstm_metrics_all_stations.csv'
)
if 'Model' not in f2.columns:
    f2['Model'] = 'LSTM'
f2['Source'] = 'LSTM Prediction'

# 2. Combine the DataFrames
df = pd.concat([f1, f2], ignore_index=True)

# 3. Select the KPI to plot
kpi = 'MAE'

# 4. Create the boxplot without hue, using Seaborn’s default palette per model
plt.figure(figsize=(10, 6))
sns.boxplot(x='Model', y=kpi, data=df, palette='Set2')

# 5. Configure titles and labels
plt.title(f'Boxplot of {kpi} by Model')
plt.xlabel('Model')
plt.ylabel(kpi)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load & tag source
f1 = pd.read_csv(
    r'C:\Users\asus\Dokumen_Billi\allstationpredgit\result\metrics_all_stations5days.csv'
)
f1['Source'] = '5days All Stations'

f2 = pd.read_csv(
    r'C:\Users\asus\Dokumen_Billi\allstationpredgit\result\lstm_metrics_all_stations.csv'
)
if 'Model' not in f2.columns:
    f2['Model'] = 'LSTM'
f2['Source'] = 'LSTM Prediction'

df = pd.concat([f1, f2], ignore_index=True)

# 2. KPI = R2
kpi = 'R2'

# 3. Hitung batas percentil (tanpa outlier)
lower, upper = df[kpi].quantile([0.05, 0.95])

# 4. Buat plot
fig, ax = plt.subplots(figsize=(8, 5))
sns.boxplot(
    x='Model', y=kpi, data=df,
    palette='Set2',
    showfliers=False,
    ax=ax
)

# 5a. Tambahkan margin 10% di atas & bawah
ax.margins(y=0.1)

# 5b. (Alternatif) Atau set ylim dengan padding manual 10%
# padding = (upper - lower) * 0.1
# ax.set_ylim(lower - padding, upper + padding)

# 6. Styling
ax.set_title(f'Boxplot of {kpi} by Model\n(outliers hidden, padded axis)')
ax.set_xlabel('Model')
ax.set_ylabel(kpi)
ax.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```
