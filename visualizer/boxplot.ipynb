```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dan tandai source (tetap bisa disimpan di df, tapi tidak dipakai untuk hue)
f1 = pd.read_csv(
    r'C:\Users\asus\Dokumen_Billi\allstationpredgit\result\metrics_all_stations5days.csv'
)
f1['Source'] = '5days All Stations'
f2 = pd.read_csv(
    r'C:\Users\asus\Dokumen_Billi\allstationpredgit\result\lstm_metrics_all_stations.csv'
)
if 'Model' not in f2.columns:
    f2['Model'] = 'LSTM'
f2['Source'] = 'LSTM Prediction'

df = pd.concat([f1, f2], ignore_index=True)

# Pilih KPI
kpi = 'MAE'

# Plot  hue di-nonaktifkan, gunakan palette seaborn default per-model
plt.figure(figsize=(10, 6))
sns.boxplot(x='Model', y=kpi, data=df, palette='Set2')
plt.title(f'Boxplot {kpi} per Model (warna seragam per model)')
plt.xlabel('Model')
plt.ylabel(kpi)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
